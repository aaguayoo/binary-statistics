```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm
```

```python
x = np.linspace(0.,0.35,1000)
pdf = norm.pdf(x, 0.18, 0.04)
pdf_norm = pdf / (pdf.sum()/len(pdf))
```

```python
fig, ax = plt.subplots(figsize=(10,8))
ax.plot(x, pdf/pdf.max(), linewidth=2.5)
ax.vlines(0.16, 0, 1.1, color="gray")
ax.vlines(0.24, 0, 1.1, color="gray")
ax.fill_between(x, pdf/pdf.max(), where=[(xi >= 0.0 and xi <= 0.16) for xi in x], color="red", alpha=0.3)
ax.fill_between(x, pdf/pdf.max(), where=[(xi >= 0.24 and xi <= 0.35) for xi in x], color="red", alpha=0.3)
ax.set_xlabel(r"$\tilde{v}$")
ax.set_ylabel(r"Probability")
ax.set_ylim(0,1.1)
ax.set_xlim(0,0.35)
plt.savefig("gauss.pdf", dpi=300, bbox_inches='tight')
plt.tight_layout()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Datos de ejemplo
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Lneas verticales
vertical_line1 = 3
vertical_line2 = 7

# Crear la figura y los ejes
fig, ax = plt.subplots()

# Graficar la curva
ax.plot(x, y, label='Curva')

# Sombrear el rea bajo la curva entre las lneas verticales
ax.fill_between(x, y, where=[(xi >= vertical_line1 and xi <= vertical_line2) for xi in x], alpha=0.3, color='gray', label='rea sombreada')

# Dibujar las lneas verticales
ax.axvline(x=vertical_line1, color='red', linestyle='--', label='Lnea 1')
ax.axvline(x=vertical_line2, color='blue', linestyle='--', label='Lnea 2')

# Aadir etiquetas y leyenda
ax.set_xlabel('Eje X')
ax.set_ylabel('Eje Y')
ax.legend()

# Mostrar la grfica
plt.show()
```
